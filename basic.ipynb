```python
# catplot
# relplot
sns.relplot(x= 'total_bill', y = 'tip',data= data)
plt.show()
```

```python
sns.relplot(x= 'total_bill', y = 'tip', hue = 'time',data= data)
plt.show()
```

```python
sns.relplot(x= 'total_bill', y = 'tip', hue = 'time',style='smoker',data= data)
plt.show()
```

```python
sns.relplot(x= 'total_bill', y = 'tip', hue = 'time',kind='scatter',row = 'time',data= data)
plt.show()
```

```python
sns.relplot(x= 'total_bill', y = 'tip', hue = 'time',style='smoker',kind='scatter',row = 'time',col = 'smoker',data= data)
plt.show()
```
